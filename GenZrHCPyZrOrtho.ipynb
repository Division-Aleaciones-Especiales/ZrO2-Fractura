# Realizamos la iteración de rotacion y guardado de archivos VASP #

```python
rotated = Zr11m20[0].copy()
```

```python
rotated.rotate(45, 'z', center=(0, 0, 0))
```

```python
plot_atoms(rotated)
```

```python
rotated = []
dang = 15  # saltos del ángulo
for ang in range(0, 91, dang):
    rotated.append(Zr11m20[0].copy())
    rotated[-1].center()
    rotated[-1].rotate(ang, 'z', center="com")
```

```python
import matplotlib.pyplot as plt 
```

```python
for i, thisatoms in enumerate(rotated):
    fig, ax = plt.subplots(1, 2)
    plot_atoms(thisatoms, ax=ax[0])
    plot_atoms(thisatoms, ax=ax[1], rotation="90x")
    
```
